### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
# Remove missing values for compensation and Age
df_age_comp = df[['Age', 'ConvertedCompYearly']].dropna()

# Convert Age to string if needed (to handle categories)
df_age_comp['Age'] = df_age_comp['Age'].astype(str)

# Group by Age, calculate median compensation
median_comp_by_age = df_age_comp.groupby('Age')['ConvertedCompYearly'].median()

# Plot
plt.figure(figsize=(10,6))
median_comp_by_age.plot(kind='line', marker='o')
plt.title('Median Yearly Compensation by Age Group')
plt.ylabel('Median ConvertedCompYearly (USD)')
plt.xlabel('Age Group')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
# Filter Age groups that fall between '25-34' and '35-44'
age_filter = ['25-34', '35-44', '45-54']  # adjust as needed, since data might have these exact age group strings

df_filtered = df_age_comp[df_age_comp['Age'].isin(age_filter)]

median_comp_filtered = df_filtered.groupby('Age')['ConvertedCompYearly'].median()

# Plot
plt.figure(figsize=(8,5))
median_comp_filtered.plot(kind='line', marker='o', color='orange')
plt.title('Median Yearly Compensation for Ages 25 to 45')
plt.ylabel('Median ConvertedCompYearly (USD)')
plt.xlabel('Age Group')
plt.grid(True)
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Drop missing values for both columns
df_exp_js = df[['WorkExp', 'JobSatPoints_6']].dropna()

# Convert WorkExp to numeric (in case some entries are strings like 'Less than 1 year')
df_exp_js['WorkExp'] = pd.to_numeric(df_exp_js['WorkExp'], errors='coerce')

# Drop rows where conversion failed (NaN)
df_exp_js = df_exp_js.dropna()

# Group by experience and calculate average job satisfaction
mean_js_by_exp = df_exp_js.groupby('WorkExp')['JobSatPoints_6'].mean()

# Plot
plt.figure(figsize=(12, 6))
mean_js_by_exp.plot(kind='line', marker='o')
plt.title('Average Job Satisfaction (JobSatPoints_6) by Experience Level')
plt.xlabel('Years of Professional Experience')
plt.ylabel('Average Job Satisfaction')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Use the correct column: 'YearsCodePro' instead of 'YearsCodingProf'
df_exp_comp = df[['YearsCodePro', 'ConvertedCompYearly']].dropna()

# Convert 'YearsCodePro' to numeric (handle entries like 'Less than 1 year')
df_exp_comp['YearsCodePro'] = pd.to_numeric(df_exp_comp['YearsCodePro'], errors='coerce')

# Drop any rows with NaN after conversion
df_exp_comp = df_exp_comp.dropna()

# Group by experience and compute median compensation
median_comp_by_exp = df_exp_comp.groupby('YearsCodePro')['ConvertedCompYearly'].median()

# Plot
plt.figure(figsize=(12, 6))
median_comp_by_exp.plot(kind='line', marker='o')
plt.title('Median Yearly Compensation by Years of Professional Coding Experience')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Compensation (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
import pandas as pd
import matplotlib.pyplot as plt

# Select the relevant columns and drop missing values
df_exp_js = df[['YearsCodePro', 'JobSatPoints_6']].dropna()

# Convert 'YearsCodePro' to numeric (handle strings like 'Less than 1 year')
df_exp_js['YearsCodePro'] = pd.to_numeric(df_exp_js['YearsCodePro'], errors='coerce')

# Drop any rows where conversion failed (NaNs)
df_exp_js = df_exp_js.dropna()

# Group by experience and compute average job satisfaction score
avg_js_by_exp = df_exp_js.groupby('YearsCodePro')['JobSatPoints_6'].mean()

# Plot
plt.figure(figsize=(12, 6))
avg_js_by_exp.plot(kind='line', marker='o', color='green')
plt.title('Average Job Satisfaction (JobSatPoints_6) by Years of Professional Coding Experience')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Average Job Satisfaction Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```
