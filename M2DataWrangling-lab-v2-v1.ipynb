### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
print("\nUnique Countries (sample):")
print(df['Country'].value_counts().head(20))
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
country_corrections = {
    'United States of America': 'United States',
    'USA': 'United States',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'UK': 'United Kingdom',
    'Russian Federation': 'Russia',
    'The Netherlands': 'Netherlands',
    'Viet Nam': 'Vietnam',
    # Add other mappings as needed
}

df['Country'] = df['Country'].replace(country_corrections)

print("\nCountries after standardization (sample):")
print(df['Country'].value_counts().head(20))
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
employment_dummies = pd.get_dummies(df['Employment'], prefix='Employment', dummy_na=True)
print("\nOne-hot encoded Employment columns sample:")
print(employment_dummies.head())
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
missing_counts = df.isnull().sum().sort_values(ascending=False)
print("\nColumns with most missing values:")
print(missing_counts.head(10))
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
median_comp = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'].fillna(median_comp, inplace=True)
print(f"\nMissing 'ConvertedCompYearly' after imputation: {df['ConvertedCompYearly'].isnull().sum()}")
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
most_freq_remote = df['RemoteWork'].mode()[0]
df['RemoteWork'].fillna(most_freq_remote, inplace=True)
print(f"Missing 'RemoteWork' after imputation: {df['RemoteWork'].isnull().sum()}")
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
import pandas as pd
import numpy as np

# Load the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Impute missing values in 'ConvertedCompYearly' with median
median_comp = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'].fillna(median_comp, inplace=True)

# Manual Min-Max scaling
min_val = df['ConvertedCompYearly'].min()
max_val = df['ConvertedCompYearly'].max()
df['CompYearly_MinMax'] = (df['ConvertedCompYearly'] - min_val) / (max_val - min_val)

print("Min-Max scaled 'ConvertedCompYearly' sample:")
print(df[['ConvertedCompYearly', 'CompYearly_MinMax']].head())

# Log-transform (add 1 to avoid log(0))
df['CompYearly_Log'] = np.log(df['ConvertedCompYearly'] + 1)

print("\nLog-transformed 'ConvertedCompYearly' sample:")
print(df[['ConvertedCompYearly', 'CompYearly_Log']].head())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
df['CompYearly_Log'] = np.log(df['ConvertedCompYearly'] + 1)
print("\nLog-transformed 'ConvertedCompYearly' sample:")
print(df[['ConvertedCompYearly', 'CompYearly_Log']].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
def map_experience_level(x):
    try:
        years = float(x)
    except:
        if isinstance(x, str):
            if 'Less than' in x:
                return 'Junior'
            elif 'More than' in x:
                return 'Expert'
            else:
                return np.nan
        else:
            return np.nan

    if years < 2:
        return 'Junior'
    elif 2 <= years < 5:
        return 'Mid'
    elif 5 <= years < 10:
        return 'Senior'
    else:
        return 'Expert'

df['ExperienceLevel'] = df['YearsCodePro'].apply(map_experience_level)
print("\nExperienceLevel distribution:")
print(df['ExperienceLevel'].value_counts(dropna=False))

# Summary
print("\nData wrangling completed.")
```
